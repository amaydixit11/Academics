```python
import random, math

def cost(s):
    c=0
    n=len(s)
    for i in range(n):
        for j in range(i+1,n):
            if s[i]==s[j] or abs(s[i]-s[j])==abs(i-j):
                c+=1
    return c

def neighbor(s):
    n=len(s)
    if random.random()<0.2:
        i,j=random.sample(range(n),2)
        t=s[:]
        t[i],t[j]=t[j],t[i]
        return t
    i=random.randrange(n)
    r=random.randrange(1,n+1)
    while r==s[i]:
        r=random.randrange(1,n+1)
    t=s[:]
    t[i]=r
    return t

def anneal(n=8,seed=None):
    if seed is not None:
        random.seed(seed)
    s=[random.randrange(1,n+1) for _ in range(n)]
    e=cost(s)
    best=s[:]
    best_e=e
    T=10.0
    Tmin=1e-4
    alpha=0.995
    it=0
    maxit=200000
    while T>Tmin and e>0 and it<maxit:
        it+=1
        s2=neighbor(s)
        e2=cost(s2)
        if e2<e or random.random()<math.exp((e-e2)/T):
            s,e=s2,e2
            if e<best_e:
                best,best_e=s[:],e
        T*=alpha
    return best,best_e,it

def print_board(s):
    n=len(s)
    for r in range(1,n+1):
        line=[]
        for c in range(n):
            line.append('Q' if s[c]==r else '.')
        print(' '.join(line))
```

```python
sol,e,it=anneal(8)
print(sol)
print(e)
print(it)
print_board(sol)
```

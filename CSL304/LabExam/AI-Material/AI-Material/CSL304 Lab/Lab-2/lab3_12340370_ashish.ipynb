**USING MST AS HEURISTIC FOR THE A\* ALGORITHM**

```python
import heapq
import math


def minmst(graph, nodes):
    if not nodes:
        return 0
    nodes = set(nodes)
    start = next(iter(nodes))
    vis = {start}
    ed = []
    for (nbr, w) in graph[start]:
        if nbr in nodes:
            heapq.heappush(ed, (w, start, nbr))
    total = 0
    while ed and len(vis) < len(nodes):
        w, u, v = heapq.heappop(ed)
        if v not in vis:
            vis.add(v)
            total += w
            for (nbr, w2) in graph[v]:
                if nbr in nodes and nbr not in vis:
                    heapq.heappush(ed, (w2, v, nbr))
    if len(vis) != len(nodes):
        return math.inf
    return total

def ham(graph, start):
    vertex=sorted(graph.keys())
    index={v: i for i, v in enumerate(vertex)}
    n=len(vertex)

    def bit(v):
        return 1 << index[v]
    
    start_mask=bit(start)
    pq=[]
    heapq.heappush(pq,(0,0,start,start_mask,[start]))
    best={}

    while pq:
        f,g,last,vis,path=heapq.heappop(pq)

        if vis == (1<<n)-1:

            for (nbr,w) in graph[last]:
                if nbr==start:
                    return path+[start]
            continue

        st=(last,vis)
        if st in best and best[st]<=g:
            continue
        best[st]=g

        for (nxt,w) in graph[last]:
            if(vis & bit(nxt))!=0:
                continue

            mask= vis | bit(nxt)
            gn=g+w


            rem=[
                v
                for v in vertex
                if (mask & bit(v)) == 0
            ]

            nodes=set(rem) | {start, nxt}
            hn = minmst(graph,nodes)

            fn = gn + hn

            heapq.heappush(pq, (fn, gn, nxt, mask, path + [nxt]))

    return None


graph = {
    0: [(1, 4), (2, 3), (7, 8)],
    1: [(0, 4), (2, 1), (3, 7)],
    2: [(0, 3), (1, 1), (3, 2), (5, 8)],
    3: [(1, 7), (2, 2), (4, 5)],
    4: [(3, 5), (5, 1), (6, 4)],
    5: [(2, 8), (4, 1), (6, 3), (7, 6)],
    6: [(4, 4), (5, 3), (7, 2)],
    7: [(0, 8), (5, 6), (6, 2)]
}

start=0
final=ham(graph,start)

if final:
    print("Hamiltonian cycle with minimum weight:")
    print(final)
else:
    print("No Hamiltonian cycle exists.")
```

**Using DFS B&B for finding the ham cycle**

```python
import math

def dfsbound(graph, start=0):
    n=len(graph)
    bcost=[math.inf]
    bpath=[[]]

    miniedge=[min(w for temp,w in graph[u]) for u in range(n)]

    def dfs(path,visited,cost):
        u=path[-1]
        newval=cost+sum(miniedge[v] for v in range(n) if v not in visited)
        if newval>=bcost[0]:
            return

        if len(path) == n:
            for v, w in graph[u]:
                if v==start:
                    total=cost+w
                    if total<bcost[0]:
                        bcost[0]=total
                        bpath[0]=path+[start]
            return

        for v, w in graph[u]:
            if v not in visited:
                dfs(path+[v],visited | {v},cost+w)


    dfs([start], {start}, 0)

    if bcost[0]==math.inf:
        return None,math.inf
    return bpath[0],bcost[0]


# same gaph liek above
graph = {
 0: [(1, 4), (2, 3), (7, 8)],
 1: [(0, 4), (2, 1), (3, 7)],
 2: [(0, 3), (1, 1), (3, 2), (5, 8)],
 3: [(1, 7), (2, 2), (4, 5)],
 4: [(3, 5), (5, 1), (6, 4)],
 5: [(2, 8), (4, 1), (6, 3), (7, 6)],
 6: [(4, 4), (5, 3), (7, 2)],
 7: [(0, 8), (5, 6), (6, 2)]
}

path, cost = dfsbound(graph, start=0)

if path:
    print("Min. Hamiltonian cycle (DFS B&B):")
    print("path:", path)
    print("cost:", cost)
else:
    print("No cycle exists.")
```
